# Clean the data
- Imputation of data with 0% to catch NaN
- Replace the %
- Convert it to type float.

```python
columns_to_convert = ['Total Rejection', 'Total Rejection Pay',
       'Total Rejection Culture', 'Fresh | Total Rejection',
       'Fresh | Rejection Pay', 'Fresh | Rejection Culture',
       'Low | Total Rejection', 'Low | Rejection Pay',
       'Low | Rejection Culture', 'Mid | Total Rejection',
       'Mid | Rejection Pay', 'Mid | Rejection Culture',
       'High | Total Rejection', 'High | Rejection Pay',
       'High | Rejection Culture']


for column in columns_to_convert:
    df[column] = df[column].fillna('0%')  # Fill NaN with '0%'

    # Step 2: Remove '%' sign
    df[column] = df[column].str.replace('%', '')  # Remove '%' sign

    # Step 3: Convert to integer
    df[column] = (df[column].astype(float)).round(2)  # Convert and round
```

# Clean the code (contd.)
- Removed 2 columns with no Total Rejection Value.
  - Justification: The data point was empty with no fact values present.
- Dropped all the rows where the values of facts were more than 100.

```python
df.drop(df[df['Total Rejection'] == 0].index, inplace=True)

for column in columns_to_convert:
  df.drop(df[df[column] > 100.00].index, inplace=True)
```

# Normalization
To consider the variance in data, and to provide a uniform outlook, with a clear picture of mean values

```python
df[columns_to_convert] = df[columns_to_convert].apply(zscore)
```
